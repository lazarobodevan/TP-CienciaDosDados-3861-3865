##Tratativa dos dados

Neste ponto, assim como as classes terapêuticas, as tarjas também podem ser unificadas, uma vez que existe mais de uma descrição para uma mesma tarja. Além disso, as variações desse campo não são pertinentes para nossa análise e, portanto, podem ser unidas.

```python
def map_tarjas(tarja):
  if tarja in ['Tarja Vermelha (*)','Tarja Vermelha(*)','Tarja  Vermelha', 'Tarja Vermelha', 'Tarja Vermelha sob restrição']:
    return 'Tarja Vermelha'
  if tarja in ['Tarja Preta(*)','Tarja  Preta']:
    return 'Tarja Preta'
  if tarja in ['Tarja Venda Livre/Sem Tarja (*)', 'Tarja Venda Livre', 'Tarja -(*)', '- (*) ','Tarja Sem Tarja']:
    return 'Sem Tarja'

df['TARJA'] = df['TARJA'].apply(map_tarjas)
```

#Regras de associação

Conforme proprosto inicialmente, temos por objetivo a previsão de preço de medicamentos a partir de algumas variáveis como tarja, classe terapêutica, laboratório e substância.
<br/>Para que consigamos fazer as regras de associação, precisamos discretizar o valor dos remédios, ou seja, transformá-los de número com ponto flutuante para string (variável categórica). Vale ressaltar que escolhemos a coluna PF sem Impostos com o objetivo prever o preço sobre o seu valor bruto, sem acréscimos de ICMS.
<br/> Dessa forma, definimos, de forma arbitrária, os labels de cada faixa de preço, sendo eles:
*   **Barato:** preço até R\$100,00
*   **Médio:** preço até R\$500,00
*   **Caro:** preço até R\$1000,00
*   **Muito caro:** preço acima de R\$1000,00



```python
def map_valor(valor):
  if(valor > 0 and valor <= 100):
    return "Barato"
  elif(valor > 100 and valor <= 500):
    return "Médio"
  elif(valor > 500 and valor <= 1000):
    return "Caro"
  elif (valor > 1000):
    return "Muito caro"

df['PF DISC'] = df['PF Sem Impostos'].apply(lambda x : map_valor(x))
df[['PF Sem Impostos', 'PF DISC']]
```

```python
#Excluindo as linhas cujos atributos PF Sem Impostos ocasionou em NaN na coluna PF DISC
df = df[df['PF DISC'].notna()]
df['PF DISC'].unique()
```

```python
df_rules = df[['SUBSTÂNCIA','LABORATÓRIO','CLASSE TERAPÊUTICA', 'TARJA','PF DISC']]
df_rules.head()
```

```python
df_transactions = pd.get_dummies(df_rules)
```

##Extraindo itemsets frequentes

Para a extração dos itemsets frequentes levamos em consideração um suporte mínimo de 0,01 (ou 1%). Isso porque grande parte das ocorrências tiveram suporte muito pequeno, e qualquer aumento no parâmetro trazia apenas algumas dezenas de transações.

```python
frequent_itemsets = apriori(df_transactions, min_support = 0.01, use_colnames=True)
frequent_itemsets
```

##Extraindo regras de associação

Neste momento extraimos as regras de associação a partir dos itemsets frequentes. Importante salientar que definimos um limite mínimo de 20% para a confiança para que mais combinações fossem trazidas.

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.2)
rules.sort_values(by="confidence", ascending=False)
```

###Extraindo métricas - Prevendo preço a partir da tarja

Aqui estimamos o preço de um medicamento tendo como base apenas a tarja, ou seja, dado uma classificação de tarja, qual o preço estimado apra aquele medicamento. Nisso, apresentamos o suporte, a confiança e lift para cada combinação de preço e tarja.

```python
def calcula_metricas(att_antecedente, att_consequente):
  print("(",att_antecedente,") -> (",att_consequente,")")

  suporte = len(df_transactions[(df_transactions[att_antecedente] == 1) & (df_transactions[att_consequente] == 1) ])/ \
  len(df_transactions)
  print("Suporte: ",suporte)

  confianca = len(df_transactions[(df_transactions[att_antecedente] == 1) & (df_transactions[att_consequente] == 1) ])/ \
  len(df_transactions[df_transactions[att_antecedente] == 1])
  print("Confiança: ",confianca)

  lift = confianca / (len(df_transactions[df_transactions[att_consequente] == 1]) / len(df_transactions))

  print("Lift: ",lift)

  print("--------------------------------")
print("##########TARJA VERMELHA##########")
calcula_metricas('TARJA_Tarja Vermelha',"PF DISC_Barato")
calcula_metricas('TARJA_Tarja Vermelha',"PF DISC_Médio")
calcula_metricas('TARJA_Tarja Vermelha',"PF DISC_Caro")
calcula_metricas('TARJA_Tarja Vermelha',"PF DISC_Muito caro")
print("\n\n##########TARJA PRETA##########")
calcula_metricas('TARJA_Tarja Preta',"PF DISC_Barato")
calcula_metricas('TARJA_Tarja Preta',"PF DISC_Médio")
calcula_metricas('TARJA_Tarja Preta',"PF DISC_Caro")
calcula_metricas('TARJA_Tarja Preta',"PF DISC_Muito caro")
print("\n\n##########SEM TARJA##########")
calcula_metricas('TARJA_Sem Tarja',"PF DISC_Barato")
calcula_metricas('TARJA_Sem Tarja',"PF DISC_Médio")
calcula_metricas('TARJA_Sem Tarja',"PF DISC_Caro")
calcula_metricas('TARJA_Sem Tarja',"PF DISC_Muito caro")
```
