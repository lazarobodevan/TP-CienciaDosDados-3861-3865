```python
df['PMC 17%'] = df['PMC 17%'].str.replace(',', '.')
df['PMC 18%'] = df['PMC 18%'].str.replace(',', '.')
df['PMC 20%'] = df['PMC 20%'].str.replace(',', '.')
df['PF 0%'] = df['PF 0%'].str.replace(',', '.')
df['PF Sem Impostos'] = df['PF Sem Impostos'].str.replace(',', '.')

df['PMC 17%'] = df['PMC 17%'].astype(float)
df['PMC 18%'] = df['PMC 18%'].astype(float)
df['PMC 20%'] = df['PMC 20%'].astype(float)
df['PF 0%'] = df['PF 0%'].astype(float)
df['PF Sem Impostos'] = df['PF Sem Impostos'].astype(float)
```

```python
def map_tarjas(tarja):
  if tarja in ['Tarja Vermelha (*)','Tarja Vermelha(*)','Tarja  Vermelha', 'Tarja Vermelha', 'Tarja Vermelha sob restrição']:
    return 'Tarja Vermelha'
  if tarja in ['Tarja Preta(*)','Tarja  Preta']:
    return 'Tarja Preta'
  if tarja in ['Tarja Venda Livre/Sem Tarja (*)', 'Tarja Venda Livre', 'Tarja -(*)', '- (*) ','Tarja Sem Tarja']:
    return 'Sem Tarja'

df['TARJA'] = df['TARJA'].apply(map_tarjas)
```

No dataset utilizado para esse trabalho, possui 3 tipos de Tarja:

*   Tarja Vermelha
*   Tarja Preta
*   Sem Tarja

Na função acima, para evitar dublicatas, o nome das três tarjas foram padronizados.




#Regras de associação

Para a regra de associação, a coluna ['PF Sem Imposto'] foi classificada da seguinte forma:



*  Para medicamentos com o preço de R\$ 0 a  99,99 reais, foi padronizado como **Barato**.
*   Para medicamentos com o preço de R$ 100,00 a 499,99 reais, foi padronizado como **Médio**.

*   Para medicamentos com o preço de R$ 500,00 a 999,99 reais, foi padronizado como **Caro**.

*  Para medicamentos com o preço a partir de R$ 1000,00 reais, foi padronizado como **Muito Caro**.

Após esses parâmetros, foi criado a coluna **PF DISC**.


```python
import sys
bins = [0, 100, 500,1000,max(df['PF Sem Impostos'])]
names = ['Barato', 'Médio', 'Caro', 'Muito caro']
df['PF DISC'] = pd.cut(df['PF Sem Impostos'], bins, labels=names)
#datasetMed.info()
df[['PF DISC','PF Sem Impostos']]
```

Para a regra de Associação, foi criado a variavel **df_rules** que irá armazenar as colunas com mais relações, sendo elas:

*   LABORATÓRIO
*   CLASSE TERAPÊUTICA

Utlizando essas colunas, foi utilizado a função **get_dummies()** que é usad para manipulação de dados. Ele converte dados categóricos em variáveis ​​fictícias ou indicadoras



```python
df_rules = df[['LABORATÓRIO','CLASSE TERAPÊUTICA', 'TARJA','PF DISC']]
df_rules.head()
```

```python
df_transactions = pd.get_dummies(df_rules)
```

```python
df_transactions.info()
```

##Extraindo itemsets frequentes

Ulizando a função apriori:

```python
frequent_itemsets = apriori(df_transactions, min_support = 0.1, use_colnames=True)
frequent_itemsets
```

Foi utilizado como suporte mínimo 0.1.

##Extraindo regras de associação

Extraindo as regras de associação, temos:

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.1)
rules.sort_values(by="confidence", ascending=False).head()
```


(PF DISC_Médio) -> (TARJA_Tarja Vermelha):
0.87 de confiança dos remedios com preço medio, possuem a tarja de cor vermelha, porem o suporte entre eles é de aproximadamente 20% e lift de 1.08.   


lift - o lift foi de 1.08, ou seja, podemos dizer que pf medio esta fortemente associado a tarja vermelha

```python
def calcula_metricas(att_antecedente, att_consequente):
  print("(",att_antecedente,") -> (",att_consequente,")")

  suporte = len(df_transactions[(df_transactions[att_antecedente] == 1) & (df_transactions[att_consequente] == 1) ])/ \
  len(df_transactions)
  print("Suporte: ",suporte)

  confianca = len(df_transactions[(df_transactions[att_antecedente] == 1) & (df_transactions[att_consequente] == 1) ])/ \
  len(df_transactions[df_transactions[att_antecedente] == 1])
  print("Confiança: ",confianca)

  lift = confianca / (len(df_transactions[df_transactions[att_consequente] == 1]) / len(df_transactions))

  print("Lif: ",lift)

  print("--------------------------------")

calcula_metricas("PF DISC_Médio","TARJA_Tarja Vermelha")
calcula_metricas("TARJA_Sem Tarja","PF DISC_Barato")
calcula_metricas("PF DISC_Barato","TARJA_Tarja Vermelha")
calcula_metricas("TARJA_Tarja Vermelha","PF DISC_Barato")
calcula_metricas("TARJA_Tarja Vermelha","PF DISC_Médio")
```
